## 4.<U>DATA CLEANING</U>:

## 6.<U>SELECTING THE APPROPRIATE MODEL</U>:
After Cleaning the Data and Exploratory Data Analysis, we will:
1. <u>Prepare the data</u>: In this, we will drop the column which has to be predicted and store this in a variable 'X'. We will initialize another variable 'y' with the column that has to be predicted( which we dropped out in the above step)
2. <u>Learning and Predicting using the MODEL</u>: In this, we will import the Logistic Regression model for predicting the 'Class' of the cancer. Logistic Regression is one of the Algorithm used in Supervised Machine learning technique. It is used for predicting the categorical dependent variable using a given set of independent variables.Therefore the outcome must be a categorical or discrete value. Hence, this model will be used in the prediction of the 'Class' of cancer. 
3. <u>Calculating the Accuracy of the MODEL</u>: In this, we will import the train_test_split module and accuracy_score module for trainig/testing and calculating the accuracy of the data , respectively. The Data Set will be divided into training(80%) and testing(20%) sets and the input values for predictions will be compared to the testing data set , which will further be compared to the testing dataset of the column to be predicted and the accuracy of the model will be displayed.

```python
#preparing the data

X = new_mydata2.drop(columns=['Class'])
y = new_mydata2['Class']
```

```python
#Learning and Predicting

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X , y)
predictions = model.predict([[4,3,3,4,6,9,2,1,2],[5,4,4,5,7,10,3,2,3],[6,5,5,6,8,9,2,3,2],[8,7,7,8,10,9,4,5,4],[9,8,8,9,10,8,5,4,5]])
predictions
```

```python
#Calculating the accuracy

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = new_mydata2.drop(columns=['Class'])
y = new_mydata2['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
score = accuracy_score(y_test, predictions)
score
```
